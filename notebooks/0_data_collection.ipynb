# Step 0 - Data collection

Process the .rar files downloaded from www.discografiascompletas.net and stores de mp3 files in drive for the next stages.

```python
import os
import shutil
import glob
from pathlib import Path
from tqdm.notebook import tqdm
from google.colab import drive

# Path with .rar files
drive_path = Path("/content/drive/MyDrive/Máster/DLASP/Final/discos")

# Folder to store the .mp3 files
drive_output_path = Path("/content/drive/MyDrive/Máster/DLASP/Final/songs")
raw_path = Path("/content/raw/")

# Password of the rar files
password = " the password "
```

```python
drive.mount('/content/drive')
```

```python
def normalize(name:str, n: int):
  """Normalizes the name of the raw mp3 files"""
  name = name.split('-')[-1]
  name = name.strip()
  name = name.lower()
  for s in ['¿']:
    name = name.replace(s, '')
  subs = {'á':'a', "é":"e", 'í': "i", "ó" : "o", "ú": "u", " " : "_", '¿': '', 
          "?": "", ",":"", "(":"", ")":"", "sr.": "sr"}
  for k, v in subs.items():
    name = name.replace(k, v)
  name = f"{n}_{name}"
  return name
```

Downloads the files and unrars them

```python
raw_path.mkdir(exist_ok=True)
os.chdir(raw_path)

discs = [rar for rar in os.listdir(drive_path) if rar.endswith('.rar')]

for disc in (pbar := tqdm(discs, leave=False)):
  pbar.set_description(disc)
  shutil.copy(drive_path / disc, raw_path / disc)
  os.system(f"unrar x {raw_path / disc} -p{password}")
  os.remove(raw_path / disc)
```

Store in drive

```python
drive_output_path.mkdir(exist_ok=True)

songs = glob.glob(f"{raw_path}/*/*.mp3")
for i, song in tqdm(enumerate(songs), leave=False):
  name = normalize(song, i+1)
  shutil.copy(song, drive_output_path / name)
```
